# Exercises using Kegg functionalities

```python
from bioservices import KEGG
k = KEGG()
```

## Get the pathway information by gene from kegg

```python
print(k.get_pathway_by_gene("aba1", "ath")) ## use aba1 if you want
```

## Visualise the pathway

```python
print(k.show_pathway("ath00906", keggid={"aba1": "red"})) ## use ath00906 pathway obtained for aba1 gene from above
```
